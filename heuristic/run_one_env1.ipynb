```python
from methods import common
import copy
from methods import ga
from methods.common import fitness,mutate
```

```python
env = common.read_env('../env/case1/0.xlsx')
fitness_value_list = ga(env,generations=100000,record=True)
```

```python
fitness_value_list
```

```python
my_dict = {}
for item in fitness_value_list:
    key = item[1]
    if key not in my_dict:
        my_dict[key] = item

result = list(my_dict.values())
print(result)
```

```python
x = [i[0] for i in result]
y = [i[1] for i in result]
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font',family='Times New Roman')
# 生成随机数据

# 绘图
fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(x, y, alpha=1, color='green', linewidth=2)
# 界限线
ax.axhline(y=30227, linestyle='-.', color='black', linewidth=1.5, label='MAML',alpha=0.8, c = 'y')
ax.axhline(y=28823.41, linestyle='--', color='black', linewidth=1.5, label='Heuristic',alpha=0.8, c = 'b')
ax.axhline(y=27457.41, linestyle=':', color='black', linewidth=1.5, label='Cplex',alpha=0.8, c = 'r')
# 标题和标签
# ax.set_title('Fitness', fontsize=28)
ax.set_xlabel('Generation', fontsize=20)
ax.set_ylabel('Fitness', fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=12)
# 图例
ax.legend(loc='best', fontsize=18)

plt.show()
```
