### 导入数据

```python
import pandas as pd
import numpy as np
import random
import matplotlib as plt
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
rawdata=pd.read_csv('./result/result_method.csv')
rawdata.columns=['SA']*5+['GA']*5+['ACO']*5+['TS']*5
### 导入十次测试的结果
data_list=[]
for i in range(5):
    data_list.append(rawdata.iloc[:,[i,i+5,i+10,i+15]])
```

### 查看大致结果

```python
import matplotlib.pyplot as plt
numbers = random.sample(range(1, 180), 10)
cols=data_list[0].columns
select=pd.concat(data_list,axis=1)
```

```python
label=range(1,6)
numbers = random.sample(range(0, 300), 10)
plt.scatter(label,select[['SA']].iloc[numbers[0]].values,label='SA')
plt.scatter(label,select[['GA']].iloc[numbers[0]].values,label='GA')
plt.scatter(label,select[['ACO']].iloc[numbers[0]].values,label='ACO')
plt.scatter(label,select[['TS']].iloc[numbers[0]].values,label='TS')
plt.xticks(label)
plt.xlabel('Times')
plt.ylabel('Obj')
plt.legend()
```

```python
# 使用 apply 方法和一个 lambda 函数将每一行最小值设置为 1，其他为 0
for i in range(5):
    data_list[i] = data_list[i].apply(lambda x: [1 if i == x.min() else 0 for i in x], axis=1, result_type='broadcast').astype(int)
```

```python
# 将两个个DataFrame相加
df=data_list[0]
for i in range(1,5):
    df=df.add(data_list[i])
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from matplotlib import rcParams
rcParams['font.family'] = 'SimHei'
 
#绘制热度图：
plt.figure(figsize=(8,6))
plot=sns.heatmap(df.head(20))
plt.title('对比')
plt.show()
```

```python
# 创建3x3子图网格
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18, 9))

# 在每个子图中绘制分布图
for i in range(2):
    for j in range(2):
        idx=(i)*2+j+1
        ax = axes[i,j]
        sns.distplot(df[100*(idx-1):idx*100].GA,label='GA',kde=True, rug=True, hist_kws={"alpha": 0.3},ax=ax)
        sns.distplot(df[100*(idx-1):idx*100].SA,label='SA', rug=True, hist_kws={"alpha": 0.3},ax=ax)
        sns.distplot(df[100*(idx-1):idx*100].ACO,label='ACO', rug=True, hist_kws={"alpha": 0.3},ax=ax)
        sns.distplot(df[100*(idx-1):idx*100].TS,label='TS', rug=True, hist_kws={"alpha": 0.3},ax=ax)
        ax.set_ylim(0,2)
        ax.set_xlabel('1st Count')
        ax.set_ylabel('Frequency')
        ax.set_title(f"Case {idx}")
        ax.grid(True)
        ax.legend()

# 调整子图之间的间距
plt.subplots_adjust(hspace=0.4, wspace=0.3)

# 显示图形
plt.show()
```
