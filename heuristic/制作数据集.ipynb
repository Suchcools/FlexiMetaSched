```python
import pandas as pd
import numpy as np
import random
import matplotlib as plt
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
rawdata=pd.read_csv('./result/result_method.csv')
rawdata.columns=['SA']*5+['GA']*5+['ACO']*5+['TS']*5
### 导入十次测试的结果
data_list=[]
for i in range(5):
    data_list.append(rawdata.iloc[:,[i,i+5,i+10,i+15]])
# 将四个DataFrame相加
df=data_list[0]
for i in range(1,5):
    df=df.add(data_list[i])
df = df.apply(lambda x: [1 if i == x.min() else 0 for i in x], axis=1, result_type='broadcast').astype(int)
df
```

```python
label = np.argmax(df.values, axis=1)
label
```

```python
import os
def find_image_file(source_path, file_lst):
    """
    递归寻找 文件夹以及子目录的 图片文件。
    :param source_path: 源文件夹路径
    :param file_lst: 输出 文件路径列表
    :return:
    """
    image_ext = ['.xlsx']
    for dir_or_file in os.listdir(source_path):
        file_path = os.path.join(source_path, dir_or_file)
        if os.path.isfile(file_path):  # 判断是否为文件
            file_name_ext = os.path.splitext(os.path.basename(file_path))  # 文件名与后缀
            if len(file_name_ext) < 2:
                continue
            if file_name_ext[1] in image_ext:  # 后缀在后缀列表中
                file_lst.append(file_path)
            else:
                continue
        elif os.path.isdir(file_path):  # 如果是个dir，则再次调用此函数，传入当前目录，递归处理。
            find_image_file(file_path, file_lst)
        else:
            print('文件夹没有环境' + os.path.basename(file_path))
env_path_list=[]
find_image_file('../env',env_path_list)
dataset = pd.DataFrame([env_path_list,label],index=['path','label']).T
dataset = dataset.sample(400)
dataset
```

```python
dataset.to_csv('../prototypical/dataset/data_info.csv',index=False)
```
