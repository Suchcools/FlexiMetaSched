### 导入数据

```python
import pandas as pd
import numpy as np
import random
import matplotlib as plt
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
rawdata=pd.read_csv('./result/result_method.csv')
rawdata.columns=['SA']*5+['GA']*5+['ACO']*5+['TS']*5
### 导入十次测试的结果
data_list=[]
for i in range(5):
    data_list.append(rawdata.iloc[:,[i,i+5,i+10,i+15]])
```

```python
# 使用 apply 方法和一个 lambda 函数将每一行最小值设置为 1，其他为 0
for i in range(5):
    data_list[i] = data_list[i].apply(lambda x: [1 if i == x.min() else 0 for i in x], axis=1, result_type='broadcast').astype(int)
```

```python
# 将两个个DataFrame相加
df=data_list[0]
for i in range(1,5):
    df=df.add(data_list[i])
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from matplotlib import rcParams
rcParams['font.family'] = 'SimHei'
 
#绘制热度图：
plt.figure(figsize=(8,6))
plot=sns.heatmap(df.head(20))
plt.title('对比')
plt.show()
```

```python
# 创建3x3子图网格
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18, 9))

# 在每个子图中绘制分布图
for i in range(2):
    for j in range(2):
        idx=(i)*2+j+1
        ax = axes[i,j]
        sns.distplot(df[100*(idx-1):idx*100].GA,label='GA',kde=True, rug=True, hist_kws={"alpha": 0.3},ax=ax)
        sns.distplot(df[100*(idx-1):idx*100].SA,label='SA', rug=True, hist_kws={"alpha": 0.3},ax=ax)
        sns.distplot(df[100*(idx-1):idx*100].ACO,label='ACO', rug=True, hist_kws={"alpha": 0.3},ax=ax)
        sns.distplot(df[100*(idx-1):idx*100].TS,label='TS', rug=True, hist_kws={"alpha": 0.3},ax=ax)
        ax.set_ylim(0,2)
        ax.set_xlabel('1st Count')
        ax.set_ylabel('Frequency')
        ax.set_title(f"Case {idx}")
        ax.grid(True)
        ax.legend()

# 调整子图之间的间距
plt.subplots_adjust(hspace=0.4, wspace=0.3)

# 显示图形
plt.show()
```

```python
df['SA'][df['SA']<4]=0
df['SA'][df['SA']>=4]=1
df['GA'][df['GA']<10]=0
df['GA'][df['GA']==10]=1
row_sum = df.sum(axis=1)
# 创建布尔索引选择需要保留的行
keep = row_sum != 0
# 保留需要的行，将其保存为新的数据帧
df_new = df[keep]
# 添加一列，值为索引+'path'
path = 'case1/'+ df_new.index.map(str) + '.xlsx'
pop = 'order_matrix/'+ df_new.index.map(str) + '.xlsx'
target = df_new.apply(lambda row: 0 if row['SA'] == 1 else 1 if row['GA'] == 1 else None, axis=1)
dataset = pd.DataFrame([path,pop,target],index=['path','pop','method']).T
```

### 制作数据集 (case 和 split)

```python
train=pd.read_csv('dataset_train.csv')
train=train[train['method'].isin([0,1])]
train['method']=train['method'].astype(int)
train
```

```python
test=pd.read_csv('dataset_test.csv')
test=test[test['method'].isin([0,1])]
test['method']=test['method'].astype(int)
test.head()
```

```python
#### 在env生成case1-4
import shutil
def copy_file(path1,method):
    idx=path1[path1.rfind('/')+1:path1.rfind('.xlsx')]
    save_path = f'env/case{method}/{idx}.xlsx'
    shutil.copy2('scene_env1/'+path1, save_path) #注意训练集测试集
    return save_path
```

```python
train_record=train.apply(lambda x:copy_file(x.path,x.method),axis=1)
pd.DataFrame(train_record).to_csv('./env/train_split.txt',header=None,index=False)
train_record
```

```python
test_record=test.apply(lambda x:copy_file(x.path,x['class'],x.method),axis=1)
pd.DataFrame(test_record).to_csv('./env/test_split.txt',header=None,index=False)
```
