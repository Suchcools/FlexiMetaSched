```python
import pandas as pd
import numpy as np
import random
import matplotlib as plt
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
rawdata=pd.read_csv('../heuristic/result/result_method.csv')
rawdata.columns=['SA']*5+['GA']*5+['ACO']*5+['TS']*5
### 导入十次测试的结果
data_list=[]
for i in range(5):
    data_list.append(rawdata.iloc[:,[i,i+5,i+10,i+15]])
# 将四个DataFrame相加
df=data_list[0]
for i in range(1,5):
    df=df.add(data_list[i])
df = df.apply(lambda x: [1 if i == x.min() else 0 for i in x], axis=1, result_type='broadcast').astype(int)
df
```

```python
df.head(500).sum(axis=0)
```

### 筛选数据

```python
select_data=[]
select_methods = ['ACO','GA','TS','SA']
for i in range(4):
    temp = df.iloc[i*500:(i+1)*500]
    sample = random.randint(85, 95)
    belone = temp[temp[select_methods[i]]==1].head(sample)
    nbelone = temp[~temp.index.isin(belone.index)].head(100-sample)
    select_data.append(pd.concat([belone,nbelone]).sort_index())
```

### 存储

```python
data = pd.concat(select_data)
data.to_csv('./索引文件.csv')
data
```

### 删除指定文件

0-499 500-999 1000-1499 1500-1999

```python
import os
for i in range(4):
    temp_pd = data.iloc[i*100:100*(i+1)]
    #去scene_env里删除别的
    dir_list = os.listdir(f'../env/case{i+1}')
    need_list = [str(x-(i)*500)+'.xlsx' for x in temp_pd.index]
    delete_list = list(set(dir_list)-set(need_list))
    [os.remove(f'../env/case{i+1}/'+x) for x in delete_list]
```

```python
import os

for ii in range(4):
    # 设置文件夹路径和新文件名前缀
    folder_path = f'../env/case{ii+1}'
    filepath = sorted(os.listdir(folder_path), key=lambda x: int(x.split('.')[0]))
    # 遍历文件夹中的所有文件
    for i, filename in enumerate(filepath):
        # 获取文件名和扩展名
        name, ext = os.path.splitext(filename)
        # 新文件名
        new_name = f'{i}{ext}'
        # 旧文件路径
        old_path = os.path.join(folder_path, filename)
        # 新文件路径
        new_path = os.path.join(folder_path, new_name)
        # 重命名文件
        os.rename(old_path, new_path)
        print(old_path,new_path)
```
