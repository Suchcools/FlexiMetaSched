```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore") 
import sys
sys.path.append('../heuristic')
from methods.common import fitness,parse_env
```

```python
def order_encodere(order):
    A=[[int(x.split('_')[0]),0,int(x.split('_')[1])] for x in order['order'][0]]
    B=[[int(x.split('_')[0]),1,int(x.split('_')[1])] for x in order['order'][1]]
    interval = 1  # 间隔
    # 将列表 B 间隔地插入到列表 A 中
    for i in range(len(B)):
        A.insert((i + 1) * interval + i, B[i])
    return A

def get_fitness(x):
    J, M, A, D, N, pt, p, W = parse_env(x['path'].replace('./','../env/'))
    return -fitness(x['bob'], J, M, A, D, N, pt, p, W)


data_path = pd.read_csv('../env/env_info.csv')
data_path['bob'] = data_path.apply(lambda x : order_encodere(np.load(f'./result/{x["class"]}_{x["excel_num"]}.npz',allow_pickle=True)),axis=1)
data_path['fitness'] = data_path.apply(lambda x: get_fitness(x),axis=1)
data_path.to_csv('../env/bob_info.csv',index=False)
```
