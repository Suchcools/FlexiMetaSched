```python
import pandas as pd
import numpy as np
```

### Case 1

```python
def gen_instance(n_job: int = 3, n_machine: int = 2, name: str = 'demo'):
    '''
    Time unit is minute
    '''
    # Prepare time in between
    W = np.repeat(np.random.randint(2, 8, size=1) *3 , n_job)
    # Amount of each job
    N = np.random.randint(2, 10, size=n_job) *3
    # Arrival time of each job
    A = np.random.randint(0, 30, size=n_job) * 3
    # Due date of each job
    D = np.random.randint(12, 22, size=n_job) * 3 *4 *2 + A
    Job = pd.DataFrame({
        "准备时间": W,
        "需求量": N,
        "到达时刻": A,
        "交货期": D
    })
    Job.index.name = 'Job ID'

    # Idle power
    P0 = np.random.randint(2, 4, size=n_machine)
    # Work power
    P1 = np.random.randint(5, 9, size=n_machine)
    Machine = pd.DataFrame({
        "闲置功率": P0,
        "工作功率": P1
    })
    Machine.index.name = 'Machine ID'

    # 1st task Process time matrix
    T1 = np.random.randint(5, 13, size=(n_job, n_machine))
    # 2nd task Process time matrix
    T2 = np.random.randint(7, 15, size=(n_job, n_machine))
    PT1 = pd.DataFrame(T1, columns=[f'Machine_{i}' for i in range(n_machine)], index=[f"Job_{i}" for i in range(n_job)])
    PT2 = pd.DataFrame(T2, columns=[f'Machine_{i}' for i in range(n_machine)], index=[f"Job_{i}" for i in range(n_job)])
    

    with pd.ExcelWriter(f'{name}.xlsx') as writer:
        Job.to_excel(writer, sheet_name='Job info')
        Machine.to_excel(writer, sheet_name='Machine info')
        PT1.to_excel(writer, sheet_name='Process 1 Time')
        PT2.to_excel(writer, sheet_name='Process 2 Time')
for i in range(500):
    gen_instance(n_job = 75, n_machine = 2, name = f'./case1/{i}')
```

### Create Env_Info

```python
import os
def find_image_file(source_path, file_lst):
    """
    递归寻找 文件夹以及子目录的 图片文件。
    :param source_path: 源文件夹路径
    :param file_lst: 输出 文件路径列表
    :return:
    """
    image_ext = ['.xlsx']
    for dir_or_file in os.listdir(source_path):
        file_path = os.path.join(source_path, dir_or_file)
        if os.path.isfile(file_path):  # 判断是否为文件
            file_name_ext = os.path.splitext(os.path.basename(file_path))  # 文件名与后缀
            if len(file_name_ext) < 2:
                continue
            if file_name_ext[1] in image_ext:  # 后缀在后缀列表中
                file_lst.append(file_path)
            else:
                continue
        elif os.path.isdir(file_path):  # 如果是个dir，则再次调用此函数，传入当前目录，递归处理。
            find_image_file(file_path, file_lst)
        else:
            print('文件夹没有环境' + os.path.basename(file_path))
env_path_list=[]
find_image_file('./',env_path_list)
labels=[int(x[x.find('case')+4:x.rfind('/')]) for x in env_path_list]
```

```python
import re
dataset=pd.DataFrame([env_path_list,labels],index=['path','class']).T
# 定义一个正则表达式来匹配路径中的case类型和excel编号
pattern = r'/case(\d+)/(\d+)\.xlsx'

# 定义一个函数来处理path列
def process_path(path):
    # 使用正则表达式匹配路径中的case类型和excel编号
    match = re.findall(pattern, path)
    # 第一个元素表示case类型，第二个元素表示excel编号
    case_type, excel_num = match[0]
    # 返回case类型和excel编号
    return int(case_type), int(excel_num)

# 调用process_path函数处理path列，将结果存储在新的列中
dataset['class'], dataset['excel_num'] = zip(*dataset['path'].apply(process_path))
dataset = dataset.sort_values(['class','excel_num'])
dataset
```

```python
dataset.to_csv('env_info.csv',index=False,encoding='utf_8_sig')
```
